```python
from absl import flags
from ml_collections import config_flags
FLAGS = flags.FLAGS

config_flags.DEFINE_config_file(
    "config_env",
    "configs/envs/metaworld.py",
    "File path to the environment configuration.",
    lock_config=False,
)

# Parsing the flags before accessing them
flags.FLAGS(["dummy_arg"])  # You need to pass at least one argument to parse the flags
```

```python
from envs.make_env import make_env
from gymnasium.vector import AsyncVectorEnv
import gymnasium as gym
import time

seed = 0
config_env = FLAGS.config_env
config_env, env_name = config_env.create_fn(config_env)

n_env_list = [1, 2, 4, 8, 16, 32, 64, 128]
d_time = {}

for n_env in n_env_list:
    env = AsyncVectorEnv([lambda i=i: make_env(env_name, seed + i, mode="train") for i in range(n_env)], 
                        autoreset_mode= gym.vector.AutoresetMode.DISABLED)

    n_step = 10000
    start_time = time.time()
    obs = env.reset()
    for _ in range(n_step):
        actions = env.action_space.sample()
        _, _, _, truncation, _ = env.step(actions)
        if truncation.any():
            obs = env.reset()
    end_time = time.time()
    d_time[n_env] = (end_time - start_time) / n_env / n_step
```

```python
for n_env in n_env_list:
    print(f"n_env: {n_env}")
    print(f"Average time per step: {d_time[n_env]} seconds")
```
